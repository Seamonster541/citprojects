# Bar Chart

```python
# Required
import matplotlib.pyplot as plt
# "Magic" notebook ommand - shows plot "inline"
%matplotlib inline

hill_heights = [44, 55, 22, 66, 78.9]

x = [1, 2, 3, 4, 5]

# create a bar chart
plt.bar(x, hill_heights, align='center')

# label the axes
plt.xlabel('Hills')
plt.ylabel('Height of the hills')

# show the bar chart
plt.show()
```

# Pie chart with label

```python
import matplotlib.pyplot as plt
%matplotlib inline

class_numbers = [42, 15, 11, 9]

class_labels = ['Fresh', 'Soph', 'Junior', 'Senior']

plt.pie(class_numbers, labels=class_labels)

plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

class_numbers = [42, 15, 11, 9]

class_labels = ['Fresh', 'Soph', 'Junior', 'Senior']

# non-zero pieces "explodes" (moves out of circle, partly)
class_explode = [0.0, 0.0, 0.0, 0.2] # only 'explode' seiors


plt.pie(class_numbers, labels=class_labels, explode=class_explode)

plt.axis('equal') # pie chart draws as a circle

plt.show()
```

# Pie chart with percentages

```python
import matplotlib.pyplot as plt
%matplotlib inline

class_numbers = [42, 15, 11, 9]

class_labels = ['Fresh', 'Soph', 'Junior', 'Senior']

# non-zero pieces "explodes" (moves out of circle, partly)
class_explode = [0.0, 0.0, 0.0, 0.1] # only 'explode' seiors

plt.pie(class_numbers, labels=class_labels, explode=class_explode, autopct='%1.1f%%')

plt.axis('equal') # pie chart draws as a circle

plt.show()
```
