```python
image = Image.open('google.jpg')
```

```python
list_of_pixels = list(image.getdata())
```

```python
def bw_pixel(pixel):
    total = sum(pixel)
    if total > 180*3:
        return(255, 255, 255)
    else:
        return (0,0,0)
```

```python
def red_pixel(pixel):
    red = pixel[0]
    green = pixel[1]
    blue = pixel[2]
    
    
    return(0, green, 0)
```

```python
new_pixel_list = []

for pixel in list_of_pixels:
    new_pixel = bw_pixel(pixel)
    
    new_pixel_list.append(new_pixel)
```

```python
image.putdata(new_pixel_list)
```

```python
def purple_pixel(pixel):
    red = pixel[0]
    green = pixel[1]
    blue = pixel[2]
    
    
    return(red, 0, blue)
```

```python
new_pixel_list = []

for pixel in list_of_pixels:
    new_pixel = purple_pixel(pixel)
    
    new_pixel_list.append(new_pixel)
```

```python
image.putdata(new_pixel_list)
```
