## Pie chart, labels on each piece

```python
import matplotlib.pyplot as plt 
%matplotlib inline

plt.axis('equal') # pie chart draws as a circle

# Data for each university class - Fresh etc:
class_numbers = [42, 15, 11, 9]

class_labels = ['Fresh', 'Soph', 'Junior', 'Senior']

plt.pie(class_numbers, labels=class_labels)

plt.show()
```

## Pie Chart - Exploded piece, percents, shadow

```python
import matplotlib.pyplot as plt 
%matplotlib inline

class_numbers = [42, 15, 11, 9]

class_labels = ['Fresh', 'Soph', 'Junior', 'Senior']

# non-zero piece "explodes" (moves out of circle, partly)
class_explode = [0.0, 0.0, 0.0, 0.15] # only 'explode' seniors

class_colors = ['lightgreen', 'lightblue', 'salmon', 'gold']

plt.figure(figsize=(8, 8))     # Opitonal - bigger plot

plt.pie(class_numbers,         # Essential - the numbers in the pie
        labels=class_labels,   # label each piece of pie
        explode=class_explode, # 'explode' a piece of the pie 
        colors=class_colors,   # custom colors
        shadow=True,           # shadow effect
        autopct='%1.1f%%')     # shows percents 

plt.axis('equal') # pie chart draws as a circle

plt.title("University Classes")

plt.show()
```

### Plot bar chart

```python
# Bar chart 
import matplotlib.pyplot as plt
%matplotlib inline

x = [1,2,3,4,5]
y = [45, 44, 25, 30, 6]

plt.bar(x,y,align='center') # A bar chart
plt.xlabel('School Class')
plt.ylabel('Students')

labels = ['Fresh','Soph', 'Junior','Senior','Other']
plt.xticks(x, labels)

## ===== Optional =======##
plt.grid(True)
## ======================##

plt.show()
```

## Plot bar chart

```python
# Required
import matplotlib.pyplot as plt
%matplotlib inline

## =========  OPTIONAL  PLOT SIZE ==============##
# Get a bigger plot using figsize
plt.figure(figsize=(8,8))
## ============================================##

hill_heights = [64, 55, 22, 166, 178.9]
hill_numbers = [1, 2, 3, 4, 5]

hill_names = ['Skinner', 'Kelly', 'College', 'Spencer', 'Pisgah']

# Create a bar chart    x            y
bar_list = plt.bar(hill_numbers, hill_heights, align='center')

## ========= OPTIONAL Colors after you get bar_list ======##
bar_list[0].set_color('gold')
bar_list[3].set_color('red')
## =======================================================##

# Add the labels to the bar chart
plt.xticks(hill_numbers, hill_names)

# Label the axes
plt.xlabel('Local Hills')
plt.ylabel('Height of hills')

# Show the bar chart
plt.show()
```

## Pandas data produces line graph

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

url_path = 'http://www.cs.uoregon.edu/Classes/17W/cis122/data/'
filename = 'yob_all.csv'
url = url_path + filename

name_history = pd.read_csv(url)
name_history.head()
```

```python
name_prompt = 'Type a female name to look for: '
typing = input(name_prompt)
who = typing.title() # 'ann' --> 'Ann'
print()

# in query, @who means the value of variable who
# note that query uses a string; entire query is within quotes

found_names = name_history.query('name == @who and gender == "F"')
print(found_names)

# x        y                        
found_names.plot('year', 'count', 
                 kind='line', 
                 title=who, 
                 figsize=(7,7),  # bigger plot
                 grid=True)      # draw grid
plt.show()
```

## pandas data as bar chart

```python
# Bar chart 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

select = 'name == @who and gender == "F" and year > 1990'
found_names = name_history.query(select)

#                  x        y
found_names.plot('year', 'count', 
                 kind='bar', 
                 title=who, 
                 figsize=(7,7),  # bigger plot
                 grid=True)

plt.show()
```
