```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# allow display of enough rows
pd.options.display.max_rows = 150

url_path = 'http://www.cs.uoregon.edu/Classes/17W/cis122/data/'
filename = 'yob_all.csv'
url = url_path + filename

name_history = pd.read_csv(url)
```

```python
name_history.head()
```

```python
name_history.tail()
```

```python
name_prompt = 'Type a name to look for: '
typing = input(name_prompt)
who = typing.title()

gender = input('Type female or male F or M ')
gender = gender.upper()[0]

found_names = name_history.query('name == @who and gender == @gender')
print(found_names)
found_names.plot('year', 'count', kind='line', title=who)
plt.show()
```

```python
popularity_promopt = 'Popularity (1 is most popular): '
typing = input(popularity_promopt)
pop = int(typing)

# in query, @pop means the value of variable pop
found_names = name_history.query('popularity == @pop and gender == "F"')\
    .sort_values('year')
print(found_names)
#found_names.plot('year', 'count', kind='line', title=who)
```

```python
name_prompt = 'Type a name o look for: '
typing = input(name_prompt)
who = typing.title()

gender = input('Type female or male F or M: ')
gender = gender.upper()[0]

found_names = name_history[name_history['name'] == who]
found_names_refined = found_names[found_names['gender'] == gender]

print(found_names_refined)
plot_title = who + ' ' + gender
found_names_refined.plot('year', 'count', kind='line', title=plot_title)

plt.show()
```

```python
name_prompt = "Type a name to look for: "
typing = input(name_prompt)
who = typing.title()
fig, ax = plt.subplots()

#in query, @who means the value of variable who
f_names = name_history.query('name == @who and gender == "F"')
print(f_names)
ax = f_names.plot(ax=ax, x='year', y='count', kind='line', title=who, color='r')

m_names = name_history.query('name == @who and gender == "M"')
print(m_names)
ax = m_names.plot(ax=ax, x='year', y='count', kind='line', title=who, color='b')

#legend
lines, temp = ax.get_legend_handles_labels()
ax.legend(lines, ['Female', 'Male'], loc='best')
```
