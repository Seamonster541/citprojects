## Filter - create a list from a larger list

```python
# 4)  10 points
# Write Python code to create an affordable_prices list
# from a list like the prices list below

# Your program should work correctly even if it had a 
# different prices list (like the one commented out)

# 'Affordable' means price is less than 50.00

# Once you have your list, have your program compute and print
# the list's total, (rounded to 2 decimals) 
# the number of items in the list
# the average price of an item in the list (rounded to 2 decimals)

prices = [56.74, 19.22, 50.00, 78.95, 167.00, 51.40, 11.22]
# prices = [22, 88, 99, 37.50, 8000.44, 19.99]


affordable_price = []
for price in prices:
    if price < 50.00:
        affordable_price.append(price)

print("The afforable: ", affordable_price)
total = round(sum(affordable_price), 2)
print("The total: $", total)
items = len(affordable_price)
print("Items: ", items)
average = round((total / items), 2)
print("The average cost per item: $", average)
```

## Find out the right category for some property taxes

```python
# 6)  10 points

# The city of Boulder wants higher property taxes on homes of the wealthy.

#  Home value < 100000  tax rate 0.002 (0.2%)
#             < 300000  tax rate 0.005 (0.5%)
#             < 700000  tax rate 0.01  (1.0%)
#             else      tax rate 0.02  (2.0%)

# Your program will ask for the value of a home, 
# then print the tax rate and the tax due for that value of a home.

# 1 point   for getting numeric value of home
#           (if you cannot get this assign a value so your program can
#             keelp going)
# 8 points  for getting right rate and tax amount
# 1 point   printing the results

# Example value = 100000   print rate 0.5 %, tax due 500.00

home_value = int(input("How much is the home? "))

if home_value < 100000:
    tax_rate = 0.002
    tax_percent = 0.2
elif home_value < 300000:
    tax_rate = 0.005
    tax_percent = 0.5
elif home_value < 700000:
    tax_rate = 0.01
    tax_percent = 1
else:
    tax_rate = 0.02
    tax_percent = 2

tax_due = round((home_value * tax_rate), 2)    
print("Rate is", tax_percent, "% tax due $", tax_due)



```

## define a function

```python
# 7)  10 points

# Define a function that given weight in pounds, returns the 
# number of grams of medicine to give.
# 
# 5 points
#   Each 100 pounds of weight should get 2 grams of medicine.
# 2 points
#   Round the amount to 1 decimal place.
# 3 points
#   Return the rounded amount to the caller

# Example, a 120 pound person should return 2.4 grams of medicine.

# Write the rest of the function definition
def grams_of_medicine(pounds):
    grams_per_lb = round((pounds * 0.02), 1)
    return grams_per_lb
    

    
# This code will check your function; don't change it:
weight = 150
grams = grams_of_medicine(weight)
print("Weight", weight, "pounds -- needs", grams, "grams of medicine")
```

## while

```python
# 8)  10 points

# Create a list of movie stars, starting from an empty list
# Repeatedly:
#    Ask for a name or "Stop"
#    As long as a name is not "Stop", add the name to a list

# Your Python program goes here:

movie_stars = []
user_input = input("Enter a movie star or 'Stop' to quit ")

while user_input != "Stop":
    movie_stars.append(user_input)
    user_input = input("Enter a movie star or 'Stop' to quit ")
    
print("Your movie stars are ", movie_stars)
```
