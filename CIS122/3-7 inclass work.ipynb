```python
from PIL import Image
```

```python
infile = 'OregonO.png'
image = Image.open(infile)

image_width = image.size[0]
image_height = image.size[1]
print(infile,
     image.format,
     image_width, 'x', image_height,
     image.mode)

#show image ( last statement in this jupyter cell)
image
```

```python
pixel_list = list(image.getdata())
new_pixels = [ ]
for rgb_pixel in pixel_list:
    red = rgb_pixel[2]
    green = rgb_pixel[1]
    blue = rgb_pixel[2]
    red = min(red, 200)
    new_pixels.append((red, green, blue))
image.putdata(new_pixels)
image
```

```python
list_of_pixels = list(image.getdata())
```

```python
new_pixel_list = []
for pixel in list_of_pixels:
    red, green, blue, = pixel
    if green > 32 and green < 200:
        green = max(green-20, 0)
    
    new_pixel_list.append((red,green,64))

image.putdata(new_pixel_list)
image
```

```python
box = (100, 100, 800, 800)
region = image.crop(box)

region = region.transpose(Image.ROTATE_180)
image.paste(region, box)
image
```
